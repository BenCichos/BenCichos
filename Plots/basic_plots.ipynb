```julia
using Plots
```

# Basic Plotting: Line Plots

```julia
x = 1:10; y = rand(10, 2)
plot(x, y)
```

```julia
# Attributes
x = 1:20; y = rand(20, 2)
plot(x, y, title = "Two Lines" , label=["line 1"  "line 2"], lw=2)

# Modifier Function
xlabel!("x values")
```

```julia
# Backend
plotlyjs()

plot(x, y, title="This is Plotted using Plotly")
```

```julia
gr()
plot(x,y, title = "This is Plotted using gr")
```

```julia
# Changing Plotting Series
plot(x, y, seriestype = :scatter, title = "Scatter Plot")
```

```julia
# Subplots
y_multi = rand(20, 4)
plot(x, y_multi, layout = (2, 2), legend = false)
```
